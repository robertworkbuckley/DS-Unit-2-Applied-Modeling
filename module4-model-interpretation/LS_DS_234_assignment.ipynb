```python
test = df[df['Year'] == 2018]
```

```python
train = df[df['Year'] <= 2017]
```

```python
train.describe(include='all')
```

```python
from sklearn.model_selection import train_test_split

train, val = train_test_split(train, stratify=train['Year'], random_state=42)

train.shape, val.shape
```

```python
target = 'Driving_Accuracy_Percentage_-_(%)'
X_train = train.drop(columns=['Driving_Accuracy_Percentage_-_(%)', 'Player_Name'])
y_train = train[target]
X_val = val.drop(columns=['Driving_Accuracy_Percentage_-_(%)', 'Player_Name'])
y_val = val[target]
X_test = test.drop(columns=['Driving_Accuracy_Percentage_-_(%)', 'Player_Name'])
y_test = test[target]
```

```python
X_train.shape, X_val.shape, X_test.shape
```

```python
col = ['Driving_Distance_-_(AVG.)', 'Driving_Distance_-_(TOTAL_DISTANCE)',
       'Driving_Distance_-_(TOTAL_DRIVES)',
       'Driving_Distance_-_All_Drives_-_(#_OF_DRIVES)',
       'Driving_Distance_-_All_Drives_-_(ALL_DRIVES)',
       'Driving_Distance_-_All_Drives_-_(TOTAL_DISTANCE)',
       'Driving_Pct._240-260_(All_Drives)_-_(%)',
       'Driving_Pct._240-260_(All_Drives)_-_(RELATIVE_TO_PAR)',
       'Driving_Pct._240-260_(All_Drives)_-_(TOTAL_DRIVES_-_OVERALL)',
       'Driving_Pct._240-260_(All_Drives)_-_(TOTAL_DRIVES_FOR_240-260)',
       'Driving_Pct._260-280_(All_Drives)_-_(%)',
       'Driving_Pct._260-280_(All_Drives)_-_(RELATIVE_TO_PAR)',
       'Driving_Pct._260-280_(All_Drives)_-_(TOTAL_DRIVES_-_OVERALL)',
       'Driving_Pct._260-280_(All_Drives)_-_(TOTAL_DRIVES_FOR_260-280)',
       'Driving_Pct._280-300_(All_Drives)_-_(%)',
       'Driving_Pct._280-300_(All_Drives)_-_(RELATIVE_TO_PAR)',
       'Driving_Pct._280-300_(All_Drives)_-_(TOTAL_DRIVES_-_OVERALL)',
       'Driving_Pct._280-300_(All_Drives)_-_(TOTAL_DRIVES_FOR_280-300)',
       'Driving_Pct._300+_(All_Drives)_-_(%)',
       'Driving_Pct._300+_(All_Drives)_-_(RELATIVE_TO_PAR)',
       'Driving_Pct._300+_(All_Drives)_-_(TOTAL_DRIVES_-_OVERALL)',
       'Driving_Pct._300+_(All_Drives)_-_(TOTAL_DRIVES_FOR_300+)',
       'Driving_Pct._<=_240_(All_Drives)_-_(%)',
       'Driving_Pct._<=_240_(All_Drives)_-_(RELATIVE_TO_PAR)',
       'Driving_Pct._<=_240_(All_Drives)_-_(TOTAL_DRIVES_-_OVERALL)',
       'Driving_Pct._<=_240_(All_Drives)_-_(TOTAL_DRIVES_FOR_<=_240)',
         'Total_Driving_-_(DISTANCE_RANK)',
       'Total_Driving_-_(EVENTS)', 'Total_Driving_-_(TOTAL)',
       'drives_rounds_ratio', 'Year']

X_train = X_train[col].replace('E', 0)
X_train = X_train[col].astype(float)
X_val = X_val[col].replace('E', 0)
X_val = X_val[col].astype(float)
X_test = X_test[col].replace('E', 0)
X_test = X_test[col].astype(float)
```

```python
X_train.columns = X_train.columns.str.replace('[<]', 'less_than')
X_val.columns = X_val.columns.str.replace('[<]', 'less_than')
X_test.columns = X_test.columns.str.replace('[<]', 'less_than')

# less_than_columns.mask('<=', 'less_than')
```

```python
X_train.shape, X_val.shape, X_test.shape
```

```python
from xgboost import XGBRegressor
from sklearn.ensemble import RandomForestRegressor
model = XGBRegressor(max_depth= 20, random_state=42, n_jobs=-1)

model.fit(X_train, y_train)
```

```python
import numpy as np
from scipy.stats import randint, uniform
from sklearn.model_selection import RandomizedSearchCV

param_distributions = { 
    'n_estimators': randint(50, 500), 
    'max_depth': [5, 10, 15, 20, None], 
    'max_features': uniform(0, 1), 
}

search = RandomizedSearchCV(
    RandomForestRegressor(random_state=42), 
    param_distributions=param_distributions, 
    n_iter=5, 
    cv=2, 
    scoring='neg_mean_absolute_error', 
    verbose=10, 
    return_train_score=True, 
    n_jobs=-1, 
    random_state=42
)

search.fit(X_train, y_train);
```

```python
print('Best hyperparameters', search.best_params_)
print('Cross-validation MAE', -search.best_score_)
model = search.best_estimator_
```

```python
import shap
```

```python
row = X_val.iloc[[0]]
explainer = shap.TreeExplainer(model)
shap_values = explainer.shap_values(row)

shap.initjs()
shap.force_plot(
    base_value=explainer.expected_value,
    shap_values=shap_values,
    features=row
)
```
